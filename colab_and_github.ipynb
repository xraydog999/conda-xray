```python
from tensorflow.keras.datasets import mnist

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()
#  or:  (x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
# You can then inspect the shapes of the loaded data
print(f"Training data shape: {x_train.shape}")
print(f"Training labels shape: {y_train.shape}")
print(f"Test data shape: {x_test.shape}")
print(f"Test labels shape: {y_test.shape}")
```

```python
# Flatten the images and create a DataFrame for a subset (e.g., x_train)
# Each row represents an image, and each column represents a pixel
# We'll use a small subset of x_train for demonstration
num_samples = 1000
x_train_flat = x_train[:num_samples].reshape(num_samples, -1)
df_mnist = pd.DataFrame(x_train_flat)
```

```python
# Print the first 5 rows
print("First 5 rows of MNIST DataFrame:")
print(df_mnist.head(5))
```

```python
# Print the last 5 rows
print("\nLast 5 rows of MNIST DataFrame:")
print(df_mnist.tail(5))
```

```python
# Display the first image in the training set
# Visualize a Single Image: Display a single image from the training set using plt.imshow().
plt.imshow(x_train[0], cmap='gray')
plt.title(f"Label: {y_train[0]}")
plt.axis('off') # Hide axes
plt.show()
```

```python
# Visualize Multiple Images: Create a grid of images to visualize several examples and their corresponding labels.
# Visualize the first 5 images from the training set
# This line creates a figure and a set of subplots. It specifies a grid of 1 row and 5 columns,
# and sets the overall size of the figure.
fig, axes = plt.subplots(1, 5, figsize=(10, 5))
# This flattens the axes array, making it easier to iterate through each subplot.
axes = axes.flatten()
# The code is not selecting images randomly. The for i in range(5): loop iterates through the numbers 0 to 4.
# This means it is selecting and displaying the first 5 images from
# the x_train dataset in the order they appear in the dataset.
for i in range(5):
  # This displays the i-th image from the training data (x_train[i]) in the
  # i-th subplot (axes[i]). cmap='gray' renders the images in grayscale.
    axes[i].imshow(x_train[i], cmap='gray')
    # This sets the title of each subplot to the corresponding label from the training labels
    axes[i].set_title(f"Label: {y_train[i]}")
    # This hides the axes for each subplot, making the image more prominent.
    axes[i].axis('off')

#  This adjusts the spacing between subplots to prevent labels and titles from overlapping.
plt.tight_layout()

# This displays the generated plot with the 5 images.
plt.show()

# Display the first image in the training set
# plt.imshow(x_train[0], cmap='gray')
# plt.title(f"Label: {y_train[0]}")
# plt.axis('off') # Hide axes
# plt.show()
```
