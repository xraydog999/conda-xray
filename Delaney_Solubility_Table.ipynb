## Define a function to check for aromaticity

### Subtask:
Create a Python function that takes a SMILES string as input and returns True if the compound is aromatic, and False otherwise, using RDKit.


**Reasoning**:
Create a Python function that takes a SMILES string as input and returns True if the compound is aromatic, and False otherwise, using RDKit.



```python
from rdkit import Chem

def is_aromatic(smiles_string):
  """
  Checks if a compound represented by a SMILES string is aromatic using RDKit.

  Args:
    smiles_string: The SMILES string of the compound.

  Returns:
    True if the compound contains any aromatic rings, False otherwise.
  """
  mol = Chem.MolFromSmiles(smiles_string)
  if mol is None:
    return False
  # Sanitize the molecule to perceive aromaticity
  Chem.SanitizeMol(mol, Chem.SanitizeFlags.SANITIZE_KEKULIZE | Chem.SanitizeFlags.SANITIZE_SETAROMATICITY)

  for atom in mol.GetAtoms():
    if atom.GetIsAromatic():
      return True
  return False

# Test the function with some examples
print(f"'c1ccccc1' is aromatic: {is_aromatic('c1ccccc1')}") # Benzene
print(f"'CCO' is aromatic: {is_aromatic('CCO')}") # Ethanol
print(f"'C1CCCCC1' is aromatic: {is_aromatic('C1CCCCC1')}") # Cyclohexane
print(f"'N#Cc1ccccc1' is aromatic: {is_aromatic('N#Cc1ccccc1')}") # Benzonitrile
print(f"'CC(C)=CCCC(C)=CC(=O)' is aromatic: {is_aromatic('CC(C)=CCCC(C)=CC(=O)')}") # Citral
```
